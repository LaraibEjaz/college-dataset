```python
import matplotlib.pyplot as plt

# Sample data
Name_C_U = ['Abilene Christian University', 'Adelphi University', 'Albion College', 'Allentown Coll. of St. Francis de Sales']
Top10perc = [23, 16, 37, 38]
Top25perc = [52, 29, 68, 23]
F_Undergrad = [2885, 2683, 1594, 113]

# Calculate total values for each university
total_values = [sum(x) for x in zip(Top10perc, Top25perc, F_Undergrad)]

# Plot the donut chart
plt.figure(figsize=(8, 8))

# Outer pie representing the donut
plt.pie(total_values, labels=None, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20c.colors)

# Inner pie representing the "donut hole"
plt.pie([1], radius=0.5, colors='white')

# Add a circle at the center to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.4, color='white', linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Add labels and title
plt.legend(labels=Name_C_U, loc="center")
plt.title('Distribution of University Statistics (Donut Chart)')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
Name_C_U = ['Abilene Christian University', 'Adelphi University', 'Albion College', 'Allentown Coll. of St. Francis de Sales']
Terminal = [78, 98, 123, 163]
Books = [23, 16, 37, 38]

# Plot the combo chart
plt.figure(figsize=(10, 6))

# Plot bars
bar_width = 0.4
x_pos = np.arange(len(Name_C_U))
plt.bar(x_pos - bar_width/2, Terminal, width=bar_width, label='Terminal', color='skyblue')
plt.bar(x_pos + bar_width/2, Books, width=bar_width, label='Books', color='lightgreen')

# Add labels and title
plt.xlabel('University')
plt.ylabel('Values')
plt.title('Combo Chart')
plt.xticks(x_pos, Name_C_U, rotation=45)
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
Name_C_U = ['Abilene Christian University', 'Adelphi University', 'Albion College', 'Allentown Coll. of St. Francis de Sales']
Top10perc = [23, 16, 37, 38]
Another_Var = [10, 20, 15, 25]  # Sample data for another variable

# Define the number of bars
num_vars = len(Name_C_U)

# Compute the angle for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make the plot close to a circle
angles += angles[:1]

# Plot the figure
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))

# Draw one axe per variable and add labels
plt.xticks(angles[:-1], Name_C_U, fontsize=10)

# Plot data
ax.plot(angles, Top10perc + Top10perc[:1], linewidth=2, linestyle='solid', label='Top10perc')
ax.plot(angles, Another_Var + Another_Var[:1], linewidth=2, linestyle='solid', label='Another Variable')

# Fill area
ax.fill(angles, Top10perc + Top10perc[:1], 'b', alpha=0.1)
ax.fill(angles, Another_Var + Another_Var[:1], 'r', alpha=0.1)

# Add legend
plt.legend(loc='upper right')

# Show the plot
plt.title('Circular Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 1 to 10 with a small step
x = np.linspace(1, 10, 100)

# Calculate y values for logarithmic functions
y_log = np.log(x)
y_log10 = np.log10(x)

# Plot logarithmic functions
plt.figure(figsize=(8, 6))
plt.plot(x, y_log, label='Natural Logarithm (ln)', color='blue')
plt.plot(x, y_log10, label='Base-10 Logarithm (log10)', color='red')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Logarithmic Functions')
plt.legend()  # Add legend to differentiate between ln and log10

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
Name_C_U = ['Abilene Christian University', 'Adelphi University', 'Albion College', 'Allentown Coll. of St. Francis de Sales']
Top10perc = [23, 16, 37, 38]
Top25perc = [52, 29, 68, 23]
F_Undergrad = [2885, 2683, 1594, 113]

# Plot the network distribution chart
plt.figure(figsize=(10, 6))

# Plot each variable for each university
bar_width = 0.25
index = range(len(Name_C_U))

plt.bar(index, Top10perc, width=bar_width, label='Top10perc')
plt.bar([i + bar_width for i in index], Top25perc, width=bar_width, label='Top25perc')
plt.bar([i + 2 * bar_width for i in index], F_Undergrad, width=bar_width, label='F.Undergrad')

# Add labels and title
plt.xlabel('University')
plt.ylabel('Values')
plt.title('Network Distribution Chart of University Statistics')
plt.xticks([i + bar_width for i in index], Name_C_U, rotation=45)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
